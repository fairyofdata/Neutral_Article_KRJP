```python
import os
import pandas as pd
from google.colab import drive

# 구글 드라이브 마운트
drive.mount('/content/drive')

# 데이터가 저장된 디렉토리 경로 설정
directory_path = '/content/drive/MyDrive/Kwargs/esg 긍부정도/'

# _positivity_filtered.csv로 끝나는 파일들을 리스트로 수집
csv_files = [f for f in os.listdir(directory_path) if f.endswith('_positivity_filtered.csv')]

# 각 파일을 데이터프레임으로 불러와서 리스트에 저장
dataframes = [pd.read_csv(os.path.join(directory_path, file)) for file in csv_files]
```

```python
# 불러온 파일 리스트 확인
print("불러온 파일 리스트:")
for file in csv_files:
    print(file)
```

```python
# 기업명 추출 함수 (파일명에서 기업명 부분만 추출)
def extract_company_name(filename):
    return filename.split('_positivity_filtered.csv')[0]

# 각 파일을 데이터프레임으로 불러와서 리스트에 저장
dataframes = []
global_index = 1  # Global ID를 위한 초기값 설정

for file in csv_files:
    # 파일 경로
    file_path = os.path.join(directory_path, file)
    # 데이터프레임 불러오기
    df = pd.read_csv(file_path)
    # 기업명 추가
    company_name = extract_company_name(file)
    df['Company'] = company_name

    # 기업별 ID 추가
    df['Company_ID'] = range(1, len(df) + 1)

    # 전체적으로 고유한 ID 추가
    df['Global_ID'] = range(global_index, global_index + len(df))
    global_index += len(df)

    # 리스트에 추가
    dataframes.append(df)

# 모든 데이터프레임을 하나로 합침
combined_dataframe = pd.concat(dataframes, ignore_index=True)

# 필요한 열만 남기고 나머지 열 제거
filtered_dataframe = combined_dataframe[['Company', 'full_text', 'Company_ID', 'Global_ID']]

# 정리된 데이터 확인
print(filtered_dataframe.head())

# 필요에 따라 파일로 저장
filtered_dataframe.to_csv('/content/drive/MyDrive/Kwargs/esg 긍부정도/filtered_company_fulltext_with_ids.csv', index=False)
```
